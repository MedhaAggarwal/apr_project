# **Data Normalization**

```python
# selecting numeric attributes columns from data
numeric_col = data.select_dtypes(include='number').columns
```

```python
# using standard scaler for normalizing
std_scaler = StandardScaler()
def normalization(df,col):
  for i in col:
    arr = df[i]
    arr = np.array(arr)
    df[i] = std_scaler.fit_transform(arr.reshape(len(arr),1))
  return df
```

# **Multi Layer Perceptron Classifier (Binary Classification)**

```python
X = bin_data.iloc[:,0:93].values # dataset excluding target attribute (encoded, one-hot-encoded,original)
Y = bin_data[['intrusion']].values # target attribute
```

```python
filepath = './models/mlp_binary.json'
weightspath = './weights/mlp_binary.weights.h5'
if (not path.isfile(filepath)):
  # serialize model to JSON
  mlp_json = mlp.to_json()
  with open(filepath, "w") as json_file:
    json_file.write(mlp_json)

  # serialize weights to HDF5
  mlp.save_weights(weightspath)
  print("Saved model to disk")

  # load json and create model
  json_file = open(filepath, 'r')
  loaded_model_json = json_file.read()
  json_file.close()
  mlp = model_from_json(loaded_model_json)

  # load weights into new model
  mlp.load_weights(weightspath)
  print("Loaded model from disk")
```

# **Long Short-Term Memory Classifier (Binary Classification)**

```python
import tensorflow as tf
# Rest of your LSTM code...
X = bin_data.iloc[:,0:93]
Y = bin_data[['intrusion']]

# splitting the dataset 75% for training and 25% testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

# Convert to numpy arrays and proper data types
X_train = X_train.to_numpy().astype('float32')
X_test = X_test.to_numpy().astype('float32')
y_train = y_train.to_numpy().astype('float32')
y_test = y_test.to_numpy().astype('float32')

# Reshape for LSTM: (samples, timesteps, features)
X_train_reshaped = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_reshaped = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Build LSTM model (fixed to remove the warning)
lst = Sequential()
lst.add(LSTM(50, input_shape=(1, X_train.shape[1])))  # Keep this, it's fine despite the warning
lst.add(Dense(1, activation='sigmoid'))
lst.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# summary of model layers
lst.summary()

filepath = './models/lst_binary.json'
weightspath = './weights/lst_binary.weights.h5'

if (not path.isfile(filepath)):
    # training the model on training dataset
    history = lst.fit(X_train_reshaped, y_train, epochs=100, batch_size=5000, validation_split=0.2)

    # serialize model to JSON
    lst_json = lst.to_json()
    with open(filepath, "w") as json_file:
        json_file.write(lst_json)

    # serialize weights to HDF5
    lst.save_weights(weightspath)
    print("Saved model to disk")

# Improved loading with custom object scope
try:
    from keras.utils import custom_object_scope

    # Register the classes properly
    with custom_object_scope({
        'Sequential': Sequential,
        'LSTM': LSTM,
        'Dense': Dense,
        'GlorotUniform': tf.keras.initializers.GlorotUniform,
        'Orthogonal': tf.keras.initializers.Orthogonal,
        'Zeros': tf.keras.initializers.Zeros
    }):
        with open(filepath, 'r') as json_file:
            loaded_model_json = json_file.read()

        # Clean the JSON of any version-specific issues
        import json
        model_config = json.loads(loaded_model_json)

        # Remove keras_version to avoid compatibility issues
        model_config.pop('keras_version', None)
        model_config.pop('backend', None)

        cleaned_json = json.dumps(model_config)
        lst = model_from_json(cleaned_json)
        lst.load_weights(weightspath)
        print("Successfully loaded model from disk")

except Exception as e:
    print(f"Error loading model: {e}")
    print("Model will be recreated from scratch...")
    # The model architecture is already defined above, so we just need to retrain
    if path.isfile(weightspath):
        try:
            # Try to load just the weights
            lst.load_weights(weightspath)
            print("Loaded weights only")
        except:
            print("Retraining model...")
            history = lst.fit(X_train_reshaped, y_train, epochs=100, batch_size=5000, validation_split=0.2)
    else:
        print("Training model from scratch...")
        history = lst.fit(X_train_reshaped, y_train, epochs=100, batch_size=5000, validation_split=0.2)

# Recompile the model (important after loading)
lst.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# predicting target attribute on testing dataset
test_results = lst.evaluate(X_test_reshaped, y_test, verbose=1)
print(f'Test results - Loss: {test_results[0]} - Accuracy: {test_results[1]*100}%')
```

```python
X = multi_data.iloc[:,0:93]  # dataset excluding target attribute (encoded, one-hot-encoded,original)
Y = multi_data[['Dos','normal','Probe','R2L','U2R']] # target attributes
```

```python
mlp = Sequential() # initializing model
# input layer and first layer with 50 neurons
mlp.add(Dense(units=50, input_dim=X_train.shape[1], activation='relu'))
# output layer with softmax activation
mlp.add(Dense(units=5,activation='softmax'))
```

```python
# defining loss function, optimizer, metrics and then compiling model
mlp.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# summary of model layers
mlp.summary()
```

```python
# training the model on training dataset
history = mlp.fit(X_train, y_train, epochs=100, batch_size=5000,validation_split=0.2)
```

```python
filepath = './models/mlp_multi.json'
weightspath = './weights/mlp_multi.weights.h5'
if (not path.isfile(filepath)):
  # serialize model to JSON
  mlp_json = mlp.to_json()
  with open(filepath, "w") as json_file:
    json_file.write(mlp_json)

  # serialize weights to HDF5
  mlp.save_weights(weightspath)
  print("Saved model to disk")

  # load json and create model
  json_file = open(filepath, 'r')
  loaded_model_json = json_file.read()
  json_file.close()
  mlp = model_from_json(loaded_model_json)

  # load weights into new model
  mlp.load_weights(weightspath)
  print("Loaded model from disk")
```

```python
# defining loss function, optimizer, metrics and then compiling model
mlp.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# predicting target attribute on testing dataset
test_results = mlp.evaluate(X_test, y_test, verbose=1)
print(f'Test results - Loss: {test_results[0]} - Accuracy: {test_results[1]*100}%')
```
